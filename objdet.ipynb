```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
config_file='C:\\Users\\91970\\Downloads\\ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model='C:\\Users\\91970\\Downloads\\frozen_inference_graph.pb'
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
class_labels=[]
file_name="C:\\Users\\91970\\Downloads\\coco.names"
with open(file_name,'rt') as f:
    class_labels=f.read().rstrip('\n').split('\n')
```

```python
print(len(class_labels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean(127.5)
model.setInputSwapRB(True)
```

```python
img=cv2.imread("C:\\Users\\91970\\Downloads\\toa-heftiba-x9I-6yoXrXE-unsplash-1024x683.jpg")
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
classIndex, confidece,bbox=model.detect(img,confThreshold=0.5)
```

```python
print(bbox)
```

```python
font_scale=3
font=cv2.FONT_HERSHEY_PLAIN
for cI,cn,bb in zip(classIndex.flatten(),confidece.flatten(),bbox):
    cv2.rectangle(img,bb,(0,0,255),2)
    cv2.putText(img,class_labels[cI-1],(bb[0]+10,bb[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=8)
```
